## 🧷 O Que é  Pré-Processamento? 
Esta componente do tratamento de dados refere qualquer processamento aplicado a um conjunto de dados antes que a estes sejam aplicados outros processamentos. É um passo essencial no processo que neste relatório será mostrado.<br><br>
Permite transformar os dados num formato mais facilmente e eficazmente processado, variando de acordo com o tipo de dados em uso e o objetivo pretendido. Alguns exemplos são:<br>
» <b><i>Sampling</i></b>, onde se separam exemplos (*samples*) de uma população larga de informação<br>
» <b><i>Denoising</i></b>, para remvoer ruído dos dados<br>
» <b>Normalização</b>, que organiza os dados de forma mais eficiente<br>
» <b>Extração de componentes</b>, que destaca características mais importantes<br><br>
<b>Mas porque é que é necessário?</b><br>
O uso de pré-processamento não é um processo apenas ocasionalmente necessário - praticamente todos os dados de escala mais robusta necessitam de algum tipo de pré-processamento. Qualquer informação que recebemos no nosso dia a dia é algo "desarrumada" - criada, passada e processada por várias pessoas, empresas e aplicações. Pode perder características, conter erros, "enfeites" desnecessários e pontos que acabam por ser redundantes. Estas imperfeições podem facilmente ser identificadas por um humano "Isso é desnecessário!", "Posso focar-me apenas nesta parte!", "Esta informação tem mais peso quando vinda de tal fonte!". Estas são conclusões a que uma máquina não pode chegar - mas que pode ser ajudada ou influenciada a fazê-lo! <br><br>
<b>E neste trabalho, o que será usado?</b><br>
Este projeto pode ser visto como uma análise de sentimento - as palavras terão um certo peso como sendo "positivas" ou "negativas" de aconrdo com a crítica a que estão associadas.<br>
Devido à diversidade de termos, palavras e interjeições presentes na maioria dos idiomas, se não houver nenhum pré-processamento dos textos, o vocabulário pode ter vários milhares de palavras, e de forma desnecessária! Uma palavra que só apareça numa crítica não vai ser um bom avaliador se uma outra é positiva ou negativa - não volta a aparecer! Para isto, os dados serão sujeitos a um processo de limpeza (que tem se ser igual para todos os documentos) que será visto seguidamente.

## 🧷 Aprendizagem Supervisionada vs Não Supervisionada
Como o nome indica, a aprendizagem supervisionada tem um "professor" ou "supervisionador". É usada para ensinar/treinar uma máquina usando dados já classificados. Ou seja - é fornecida à máquina uma série de informações já divididas por classe, ou com a "resposta correta". Depois disso, é-lhe fornecido um novo conjunto de dados para serem analisados e classificados da mesma forma. A máquina comparará as características das novas amostras àquelas que já conhecia, conseguindo, assim, associar-lhe, por base nisso, a classe onde melhor se enquadrará.<br><br>
**Pode ser dividida em:**<br>
» <b>Classificação:</b> Um problema de classificação refere aquele em que a variável de *output* é uma categoria ("Azul, "Vermelho")<br>
» <b>Regressão:</b> Um problema de regressão refere quando a variável de *output* é um valor real ("euros", "peso").


**Tipos:**<br>
» Regressão<br>
» Regressão Logística<br>
» Classificação<br>
» K-Vizinhos<br>
» Árvores de Decisão<br>

Tem o lado positivo de permitir optimizar o critério de decisão com base na experiência passada, solucionando muitos problemas do mundo real. Por outro lado, a classificação de dados de larga escala pode ser um desafio, precisando de muito tempo e carga computacional.
<br><br><br>
Já a **não supervisionada**, trabalha com a informação sem ser ensinada. A sua tarefa principal é agrupar informação de acordo com parecenças, padrões e diferenças sem ter antes tido acesso a dados do mesmo tipo.<br>Por exemplo, sendo dada uma imagem de laranjas e morangos, a máquina não poderá dizer "isto é um morango e isto é uma laranja", mas poderá dizer "este grupo tem características semelhantes - é em forma de gota, tem vários pontos no seu interior e uma série de folhas a sair do topo". É esperado que consiga fazer as suas próprias descobertas - trata dados sem <i>label</i>. São exemplos o ***Clustering*** e a ***Associação***.<br><br>
Neste trabalho, tratar-se-á um problema de classificação (supervisionada) tanto binária como multi-classe.

## 🧷 Funções e Variáveis Auxiliares
Nesta secção, apresenta-se uma série de funções auxiliares ao processo efetuado no seguimento do projeto.\
Cada uma inclui uma breve descrição relativa ao seu objetivo e importância/relevância no processo.

```python
# Variáveis Globais Auxiliares
imdb = 'imdbFull.p' # Nome do ficheiro que inclui as críticas usadas no treino dos classificadores
```

```python
# ↪ openFile()
#    Recebe o nome de um ficheiro que contenha informação a ser convertida/lida como pickle.
#       Argumentos:
#       » filename - Nome do ficheiro a converter;
#       Returns:
#       » Conteúdo convertido;
def openFile(filename):
    return pickle.load(open(filename, 'rb'))


# ↪ limpar()
#    Recebe um array com um ou mais documentos, limpando-os de forma a conter apenas em caracteres
#    alfabéticos sem acentos.
#       Argumentos:
#       » documentos - Array com o(s) documento(s) a limpar;
#       » binary     - flag, por omissão a 0, que indica necessidade de descodificação binária
#                      (se o ficheiro estivesse nesse formato);
#       Returns:
#       » Array com os documentos limpos;
def limpar(documentos, binary=False):
    # Passo 1 » Converter strings binárias para strings normais
    if binary:
        documentos = [doc.decode('UTF-8') for doc in documentos]
        
    # Passo 2 » Remoção de marcas html de mudança de linha
    documentos = [doc.replace('<br />', ' ') for doc in documentos]
    
    # Passo 3 » Eliminar todos os caracteres que não sejam alfabéticos
    documentos = [re.sub(r'[^a-zA-Z\u00C0-\u00FF]+' , ' ', doc) for doc in documentos]
    
    return documentos
```

## 🧷 Funções e Variáveis Auxiliares - Requisitadas no Enunciado
De seguida, mostram-se algumas funções obrigatórias conforme estipuladas no enunciado:
- ***text2vector:***  \
**» Pedido -** Função que converte uma lista de *strings* (variável *Docs*) na representação *tf-idf*. Deve retornar uma matriz documento/termo (matriz *X*). A função deverá fazer as limpezas necessárias aos documentos (*strings*) e usar um modelo *tf-idf* previamente treinado. A matriz será, depois, usada nas funções de classificação binária e multi-classe.\
**» Implementação -** A função recebe a lista de *strings*, na referida variável *Docs*, começando por limpá-la (chamando a respetiva função auxiliar). Depois, mediante o tipo de classificação que se pretende obter - algo indicado pelo argumento *tipo_classificacao* - obterá, então, de um ficheiro guardado, o modelo *tf-idf* já treinado com os melhores parâmetros encontrados, convertendo, assim, os documentos na matriz *X* pedida, retornando-a.

```python
# ↪ text2vector()
#    Recebe uma lista de strings, limpando-as e convertendo-as na matriz X, conforme modelo tf-idf
#   previamente treinado.
#       Argumentos:
#       » Docs               - Lista de string;
#       » tipo_classificacao - 'b' se binária, 'm' se multi-classe;
#       » binary             - flag, por omissão a 0, que indica necessidade de descodificação
#                              binária (se o ficheiro estivesse nesse formato);
#       Returns:
#       » Matriz X na representação tf-idf;
def text2vector(Docs, modo = 'b', binary=False):
    Docs  = limpar(Docs, binary)
    tfidf = None
   
    tfidf = openFile('tfidf_binario.p')['tfidf'] if modo == 'b' else openFile('tfidf_multiclasse.p')['tfidf']
        
    if (tfidf):
        X = tfidf.transform(Docs)
        return X
```

- ***binClassify:***  \
**» Pedido -** Função que classifica a matriz *X* da função anterior em positivos e negativos. Deve retornar um array, *y*, de 0s e 1s com o resultado da classificação. Pode usar um classificador que achar adequado, mas este já deve estar treinado.\
**» Implementação -** O seu funcionamento é bastante simples - obtém, de um ficheiro já guardado, o classificador já treinado relativo à classificação binária, retornando a previsão/classificação da matriz *X*.

```python
# ↪ binClassify()
#    Classifica a matriz X em críticas positivas e negativas.
#       Argumentos:
#       » X                            - Matriz na representação tf-idf;
#       » nClassificador               - Número do classificador a usar - por default usa-se o Linear SVC;
#       Returns:
#       » Classificação binária;
def binClassify(X, nClassificador = 2):
    classificador = None
    if nClassificador == 1:
        classificador = openFile('modelo_logistic_regression.p')['modelo']
        
    if nClassificador == 2:
        classificador = openFile('modelo_linear_svc_binario.p')['modelo']
        
    if nClassificador == 3:
        classificador = openFile('modelo_knn_binario.p')['modelo']
        
    if nClassificador == 4:
        classificador = openFile('modelo_multinomialNB_binario.p')['modelo']
    return classificador.predict(X)
```

- ***multiClassify:***  \
**» Pedido -** Função que classifica a matriz *X* devolvida pela função *text2vector* numa de oito classes (1-4 e 7-10). Deve retornar um array, *y*, de inteiros (1-4 e de 7-10) com o resultado da classificação. Pode usar um classificador que achar adequado, mas este já deve estar treinado.\
**» Implementação -** O seu funcionamento é bastante simples - obtém, de um ficheiro já guardado, o classificador já treinado relativo à classificação binária, retornando a previsão/classificação da matriz *X*.

```python
# ↪ multiClassify()
#    Classifica a matriz X nas classes 1-4 e 7-10.
#       Argumentos:
#       » X                                - Matriz na representação tf-idf;
#       » nClassificador                   - Número do classificador a usar;
#       Returns:
#       » Classificação multiclasse;
def multiClassify(X, nClassificador = 1):
    classificador = None
    if nClassificador == 1:
        classificador = openFile('modelo_logistic_regression_multiclasse.p')['modelo']
        
    if nClassificador == 2:
        classificador = openFile('modelo_linear_svc_multiclasse.p')['modelo']
        
    if nClassificador == 3:
        classificador = openFile('modelo_knn_multiclasse.p')['modelo']
        
    if nClassificador == 4:
        classificador = openFile('modelo_multinomialNB_multiclasse.p')['modelo']
    return classificador.predict(X)
```

### ◽ 1.2 Para que Serve a Vetorização? - Estudo Preliminar do Tamanho do Vocabulário
**- O que é *TF-IDF*?**\
*Term frequency-inverse document frequency* é uma forma que permite medir e quantificar a importância que dada palavra tem numa coleção de documentos (o *corpus*).\
\
***» TF (Term Frequency):*** Relaciona-se a frequência de cada termo em estudo com cada documento do seu geral. A frequência pode ser calculada na sua forma mais básica, apenas por contagens, ou tratar, ser mais ajustada/normalizada ao tamanho do documento, ou calculada com métodos mais robustos (como de forma logaritmica).\
***» IDF (Inverse Document Frequency):*** Tem em conta o quão comum (ou incomum) uma palavra é em todo o *corpus*. A importância do *IDF* assenta na correção de palavras que aparecem com muita frequência na língua em que o texto está escrito - deve-se maximizar o peso de termos com menos frequência.\
\
Quanto mais alto for a pontuação/score *TF-IDF*, mais importante/relevante é esse termo (aproximando-se de 0 quanto menos relevante for).
\
Verifique-se como o ajuste de parâmetros afetam os dados que temos em mão.\
Para isso, criam-se listas de vários valores de *min_df* e *token_pattern* a testar:
- ***min_df:*** *Minimum Document Frequency*, indica a frequência mínima com que uma palavra tem de aparecer de modo a ser considerada parte do vocabulário.
- ***token_pattern:*** Expressão regular que indica um "padrão" que uma palavra deve ter para ser considerada um *token*. Nos exemplos testados, o número de '\ws' indica o número mínimo de caracteres requisitado.

### ◽ 2.1 Definição, Uso e Metodologia
O primeiro classificador utilizado neste projeto é o discriminante logístico, que em *sklearn* (apesar de se tratar de um classificador e não se um regressor) tem o nome de *LogisticRegression*.
\
Este classificador consiste numa transformação linear, seguida de uma não-linearidade e é usado para estimar um conjunto de valores, tendo como dados disponíveis os dados de entrada e as saídas desejadas.\
\
O discriminante logístico é dado pela função $y = φ(W^T x)$, sendo que o valor mais alto de $y$ traduz a classe escolhida pelo classificador.\
\
É de salientar ainda o facto deste classificador variar de acordo com os parâmetros introduzidos, nomeadamente *tol* (Tolerância), *max_iter* (Número Máximo de Iterações), *solver* (Métodos de Otimização) e *multi_class* (difere entre classificação binária e multi-classe).\
\
O discriminante logístico também costuma ter um termo de regularização - em *sklearn* o termo - *penalty* - por omissão costuma ser a regularização *Ridge* (acrescenta-se a uma função de custo, outra função de custo - penalizar pesos com valores de 𝑊 altos), sendo também opção a regularização *Lasso* (esta vai permitir pôr vários dos valores dos pesos a 0 - se o vetor de dados (𝑥) tem uma dimensão muito elevada, ter vários dos valores a 0 permitirá escolher quais das dimensões são úteis e quais não são) e também a opção da regularização *Elastic-Net* que é, basicamente, uma combinação das regularizações *Ridge* e *Lasso*. Caso não se deseje ter nenhum termo de regularização, coloca-se o parâmetro *penalty = None*. A regularização deve ser usada quando o modelo tem a capacidade de entrar em sobre aprendizagem.\
\
As vantagens deste classificador passam pelo facto deste ter bons resultados quando o conjunto de dados é linearmente separável, ser menos propenso a sobre aprendizagem, mas poder sobre aprender em conjuntos de dados de alta dimensão - através dos termos de regularização. O discriminante logístico é ainda fácil de implementar e interpretar e muito eficiente no treino.\
\
Já as desvantagens passam pelo facto de, no mundo real, os dados raramente serem linearmente separáveis e não se verificar linearidade entre a variável dependente e as variáveis independentes. Para além disso, este classificador tem a sua utilização limitada a problemas cujo conjunto de dados é composto por números discretos, não podendo ser utilizado para estimar dados contínuos.

### ▪ 2.2.1 Procura dos Melhores Parâmetros
O próximo passo será encontrar os melhores valores relativos a ***document_frequency, token_pattern, n_grams, C*** e ***Penalty***. Falaremos mais dos *n_gramas* de seguida, mas apresenta-se agora uma função ***teste_binario*** que:<br><br>
» **Divide os documentos em treino, validação e teste** - Esta divisão permite uma averiguação menos "brusca" dos resultados.\
» **Recebe uma combinação de *document_frequency, token_pattern* e *n_grams*, usando-a para criar o *TDIDF Vectorizer*** a utilizar.\
» Transforma os documentos com essa função de vetorização.\
» Procura os pares de *C* e *Penalty* ideiais, de acordo com 4 parâmetros: **melhor validação cruzada, menor número de erros, melhor precisão, melhor score.**\
» **Retorna a média dos scores obtidos com cada um desses pares**, a fim de serem posteriormente comparados de modo a escolher os melhores parâmetros.\
\
**Validação Cruzada? Precisão?**\
A validação cruzada refere uma técnica que procura descobrir quão bem um dado modelo se generaliza para um *dataset* independente. É uma técnica usada para avaliar este tipo de modelos e estimar o quão adaptáveis estes serão.\
\
Relativamente à precisão, esta é outra métrica muitas vezes usada pra quantificar a performance de um modelo. Também chamado de valor preditivo positivo, é definido pelo número de Verdadeiros Positivos dividido pelo número de Positivos Total - ou seja, quantifica a percentagem de previsões corretamente identificadas como positivas.

```python
# Função auxiliar que mostra resultados obtidos da calibração
def avalResultados(criterio, X1t, X1v, X2, y1t, y1v, y2, par):
    print('\n    ↬ Critério:', criterio)
    lr = LogisticRegression(penalty=par[0], solver='liblinear',C=par[1], max_iter=5000, tol=1e-2)
    lr.fit(X1t, y1t)
    
    scoreTrain = lr.score(X1t, y1t)*100
    scoreVal   = lr.score(X1v, y1v)*100
    scoreTest  = lr.score(X2, y2)*100
    print('        » Acertos (treino):    ', scoreTrain)
    print('        » Acertos (validação): ', scoreVal)
    print('        » Acertos (teste) :    ', scoreTest)
    print('        » Parâmetros (cv) : penalty='+par[0]+', C=', par[1])
    
    print('        » Matriz de Confusão (teste)\n', confusion_matrix(y2, lr.predict(X2)))
    return scoreVal # Retorna a média dos scores

def teste_binario(documents, targets, mdf, tp, ngrams):
    print('*. Início de Novo Teste')
    print('    ↬ min_df:', mdf, '  token_pattern:', tp, '  n_grams:', ngrams)
    
    Docs1, Docs2, y1, y2     = sklearn.model_selection.train_test_split(documents, targets, test_size=1/3, stratify=targets)
    Docs1t, Docs1v, y1t, y1v = sklearn.model_selection.train_test_split(Docs1, y1, test_size=1/2, stratify=y1)
    
    print('\n    ⟲ Vetorização... ')
    tfidf    = TfidfVectorizer(min_df=mdf, token_pattern=tp, ngram_range=ngrams).fit(Docs1t)
    print('    ✔ Vetorização concluída! ')

    tf_info  = {'tfidf' : tfidf}
    
    pickle.dump(tf_info, open('testes_tf_idf/tfidf_test_' + str(mdf) + '_' + str(tp.replace('\\', '').replace('b', '').count('w')) + '_' + str(ngrams) + '.p', 'wb'))
    
    tokens   = tfidf.get_feature_names()
    voc_size = len(tokens)
    print('    ↬ Tamanho Vocabulário:', str(voc_size))

    X1, X1t, X1v, X2   = tfidf.transform(Docs1), tfidf.transform(Docs1t), tfidf.transform(Docs1v), tfidf.transform(Docs2)
    
    cList = [0.01, 0.1, 1, 5, 10]
    pList = ['l1', 'l2']
    
    parCVS      = parErros = parPrecision = parScore = None
    menorNErros = 50000
    melhorCVS   = 0
    melhorScore = 0
    melhorDist  = 1
    melhorPonto = None
    print('\n    ⟲ Início da procura dos melhores parâmetros... ')
    for p in pList:
        for c in cList:
            lr           = LogisticRegression(penalty=p, solver='liblinear', C=c, max_iter=5000, tol=1e-2)
            
            # Avaliação com base no cross_val_score
            cvs          = np.mean(cross_val_score(lr, X1, y1, cv=5))
            
            # Avaliação por decision function
            lr.fit(X1t, y1t)
            y1ve         = lr.predict(X1v)
            e            = np.sum(y1v != y1ve)
            
            # Score
            score        = lr.score(X1v, y1v)
            
            # Precisão
            prec, rec, _ = precision_recall_curve(y1v, y1ve)
            pontos       = np.array([prec, rec])
            dists        = np.sqrt(np.sum((pontos - np.array([[1.0], [1.0]]))**2, axis=0))
            idx_min      = np.argmin(dists)
            dist         = dists[idx_min]
            ponto_ideal  = pontos[:, idx_min]
                
            if dist <= melhorDist:
                melhorPonto  = ponto_ideal
                parPrecision = (p, c) 
            if score >= melhorScore:
                melhorScore  = score
                parScore     = (p, c) 
            if e <= menorNErros:
                menorNErros  = e
                parErros     = (p, c)
            if cvs >= melhorCVS:
                melhorCVS    = cvs
                parCVS       = (p, c)
    print('    ✔ Melhores parâmetros encontrados!')
       
    score1 = avalResultados('Melhor score', X1t, X1v, X2, y1t, y1v, y2, parScore)
    score2 = avalResultados('Menor número de erros', X1t, X1v, X2, y1t, y1v, y2, parErros)
    score3 = avalResultados('Melhor cross-validation-score', X1t, X1v, X2, y1t, y1v, y2, parCVS)
    score4 = avalResultados('Melhor valor de precisão', X1t, X1v, X2, y1t, y1v, y2, parPrecision)
    
    media  = np.mean(np.array([score1, score2, score3, score4]))
    
    print('\n    ↬ Avaliação final:', media)
    print('\n\n')
    return (voc_size, media)
```

### ▪ 2.3.1 Procura dos Melhores Parâmetros
Repete-se o processo de procura dos melhores parâmetros para o classificador multi-classe...

```python
def teste_multiclasse(documents, targets, mdf, tp, ngrams):
    print('*. Início de Novo Teste')
    print('    ↬ min_df:', mdf, '  token_pattern:', tp, '  n_grams:', ngrams)
    
    Docs1, Docs2, y1, y2     = sklearn.model_selection.train_test_split(documents, targets, test_size=1/3, stratify=targets)
    Docs1t, Docs1v, y1t, y1v = sklearn.model_selection.train_test_split(Docs1, y1, test_size=1/2, stratify=y1)
    
    print('\n    ⟲ Vetorização... ')
    tfidf    = TfidfVectorizer(min_df=mdf, token_pattern=tp, ngram_range=ngrams).fit(Docs1t)
    print('    ✔ Vetorização concluída! ')

    tf_info  = {'tfidf' : tfidf}
        
    tokens   = tfidf.get_feature_names()
    voc_size = len(tokens)
    print('    ↬ Tamanho Vocabulário:', str(voc_size))

    X1, X1t, X1v, X2   = tfidf.transform(Docs1), tfidf.transform(Docs1t), tfidf.transform(Docs1v), tfidf.transform(Docs2)
    
    cList = [0.01, 0.1, 1, 5, 10]
    pList = ['l1', 'l2']
    
    parCVS      = parErros = parPrecision = parScore = None
    menorNErros = 50000
    melhorScore = 0
    melhorPonto = None
    print('\n    ⟲ Início da procura dos melhores parâmetros... ')
    for p in pList:
        for c in cList:
            lr           = LogisticRegression(penalty=p, solver='saga', C=c, max_iter=5000, tol=1e-2, multi_class='multinomial')
            
            # Avaliação por decision function
            lr.fit(X1t, y1t)
            y1ve         = lr.predict(X1v)
            e            = np.sum(y1v != y1ve)
            
            # Score
            score        = lr.score(X1v, y1v)
                            
            if score >= melhorScore:
                melhorScore  = score
                parScore     = (p, c) 
            if e <= menorNErros:
                menorNErros  = e
                parErros     = (p, c)
    print('    ✔ Melhores parâmetros encontrados!')
       
    score1 = avalResultados('Melhor score', X1t, X1v, X2, y1t, y1v, y2, parScore)
    score2 = avalResultados('Menor número de erros', X1t, X1v, X2, y1t, y1v, y2, parErros)
    
    media  = np.mean(np.array([score1, score2]))
    
    print('\n    ↬ Avaliação final:', media)
    print('\n\n')
    return (voc_size, media)
```

## 🧷 3. Classificador n.º2 - Support Vector Machine (SVM)

### ◽ 3.1 Definição, Uso e Metodologia
As Máquinas de Suporte Vetorial (SVM - *Support Vector Machine*), são algoritmos supervisionados de aprendizagem automática aplicáveis a problemas de classificação e regressão, podendo até algumas ser adaptadas a métodos não supervisionados.\
\
No contexto de classificação, Máquinas de Suporte Vetorial não são nada mais do que classificadores binários e são, em muitos aspetos, muito parecidos com discriminantes lineares.\
\
Comparativamente com o discriminante logístico, as SVM irão separar as classes em análise de modo a que a margem entre estas seja a maior possível.\
\
Os vetores que estão mais próximos da margem das classes são os vetores de suporte, definindo a margem de separação, sendo só estes que serão usados para a classificação - os restantes pontos de treino não serão necessários. É de destacar que quantos mais vetores de suporte existirem, mais lento será todo este processo.\
\
As SVM são bons classificadores, ou seja, têm bons resultados em problemas de classificação, entre eles problemas com dados de alta dimensão, como é o caso dos dados em análise neste projeto.\
\
No entanto este tipo de classificador também possui as suas desvantagens. Para além de ser muito sensível ao escalamento dos dados, é também necessário definir uma função de *kernel*, a qual tem de ser adaptada a cada problema, não se sabendo à partida qual é o melhor *kernel* a utilizar, sendo preciso realizar testes. Uma outra desvantagem é o facto de não lidar muito bem com uma quantidade muito muito elevada de dados.

### ◽ 4.1 Definição, Uso e Metodologia
O classificador dos k vizinhos mais próximos (K-Nearest Neighbors) é um algoritmo supervisionado que realiza a classificação com base na proximidade de um dado ponto aos seus vizinhos, partindo do pressuposto que pontos semelhantes podem ser encontrados perto uns dos outros.\
\
O KNN é chamado de algoritmo preguiçoso uma vez que não realiza nenhum treino quando lhe são fornecidos os dados de treino. Em vez disso, apenas armazena os dados durante o tempo de treino e não realiza nenhum cálculo, não criando nenhum modelo até que sejam consultados os dados.\
\
Este classificador possui assim as vantagens de ter uma fácil implementação graças à sua simplicidade e precisão. É também facilmente adaptável, uma vez que à medida que os dados de treino vão sendo adicionados, o algoritmo ajusta-se para considerar quaisquer novos dados, uma vez que estes são armazenados na memória. E, por fim, uma outra vantagem deste classificador é o facto de ter poucos hiperparâmetros, precisando apenas do valor de k e de uma métrica de distância.\
\
No entanto, o KNN também possui algumas desvantagens. Nomeadamente, não funciona bem com elevados conjuntos de dados, uma vez que requer um alto armazenamento de memória, nem com dados de dimensões elevadas. É, também, necessário determinar o valor de K de acordo com o problema. Para além disso, o processo de estimação é lento se o número de vizinhos for elevado.\
\
Com isto em mente, **podem-se esperar, desde já, piores resultados do que os obtidos com os classificadores anteriores**.

O algoritmo Multinomial Naive Bayes é um método de aprendizagem probabilística que é usado principalmente em Processamento de Linguagem Natural (NLP - Natural Language Processing).\
\
Este algoritmo é baseado no teorema de Bayes (que calcula a probabilidade de um evento ocorrer com base no conhecimento prévio das condições relacionadas a um evento) e prevê a "etiqueta" de um texto, no contexto deste projeto, uma crítica a um filme.\
\
O Multinomial Naive Bayes calcula a probabilidade de cada "etiqueta" para um determinado conjunto de dados e, em seguida, fornece a "etiqueta" com a maior probabilidade como saída.\
\
Tal como todos os classificadores, o Multinomial Naive Bayes possui vantagens e desvantagens. As vantagens consistem no facto de ser de fácil implementação, dado que só é necessário calcular a probabilidade. É simples e pode ser usado para estimar aplicações em tempo real. E é altamente escalável, podendo lidar facilmente com elevados conjuntos de dados.
Quanto às desvantagens, destaca-se o facto da precisão da previsão ser menor comparativamente a outros algoritmos de propabilidade e também o facto de não ser adequado a regressão, dado que é usado apenas para classificação de dados textuais e não pode ser usado para prever valores numéricos.

## 🧷 2. PCA
<a id='section7-2-1'></a>
### ◽ 2.1 O que trata?
O PCA tem como objetivo projetar dados nas direções de maior variância (as componentes principais), sendo os dados projetados descorrelacionados (a matriz de covariância é diagonal). Pressupõe-se, assim, que direções onde os dados variam mais contêm mais informação.\
\
Consiste então num método não supervisionado - não há classes - que reduz a dimensionalidade de maneira a preservar o mais possível a variação presente nos dados de alta dimensão.\
\
**LDA - Análise em Discriminantes Lineares**\
O LDA, também conhecido por TruncatedSVD, tem como objetivo encontrar uma projeção de modo a maximizar a vâriancia inter-classe (entre classes) e minimizar variância intra-classe (dentro da mesma classe).\
\
**PCA vs LDA**\
Comparando PCA e LDA destaca-se que no PCA as componentes principais podem não ser as direções mais discriminativas
e que os dados projetados podem ter as mesmas dimensões dos dados originais. Já o LDA possui um melhor poder discriminativo e projeta os dados com c-1 dimensões, sendo c o número de classes.\
\
Verifiquemos como afeta este tipo de dados.

<a id='section7-2-2'></a>
### ◽ 2.2 Procura de Componentes Ideais - Binário
À luz dos métodos de calibração até agora utilizados, testa-se a função TruncatedSVD com vários números de componentes, criando uma função auxiliar que retorna os resultados da regressão logística após tratados com o TruncatedSVD.

```python

def test_components(comps, X1t, X1v, y1t, y1v):
    print('\n\nComponentes » ', comps)
    svd = TruncatedSVD(n_components=comps, n_iter=5)
    svd.fit(X1t)
    X1tSVD, X1vSVD = svd.transform(X1t), svd.transform(X1v)

    logReg = LogisticRegression(solver='liblinear', penalty='l2', C=3.5, max_iter=5000, tol=1e-2)
    
    logReg.fit(X1tSVD, y1t)
    
    scoret = logReg.score(X1tSVD, y1t)*100
    scorev = logReg.score(X1vSVD, y1v)*100
     
    print('Acertos (treino):    ', scoret)
    print('Acertos (validação): ', scorev)
    
    return (scoret, scorev)

def test_components_multiclasse(comps, X1t, X1v, y1t, y1v):
    print('\n\nComponentes » ', comps)
    svd = TruncatedSVD(n_components=comps, n_iter=5)
    svd.fit(X1t)
    X1tSVD, X1vSVD = svd.transform(X1t), svd.transform(X1v)

    logReg = LogisticRegression(solver='saga', penalty='l2', C=2.5, max_iter=5000, tol=1e-2, multi_class='multinomial')
    
    logReg.fit(X1tSVD, y1t)
    
    scoret = logReg.score(X1tSVD, y1t)*100
    scorev = logReg.score(X1vSVD, y1v)*100
     
    print('Acertos (treino):    ', scoret)
    print('Acertos (validação): ', scorev)
    
    return (scoret, scorev)
```

## 🧷 3. ***Clustering***
<a id='section7-3-1'></a>
### ◽ 3.1 O que trata?
Os métodos de agrupamento, ou ***clustering***, são técnicas de aprendizagem não supervisionada onde se procura encontrar representações mais compactas dos dados sem mais nenhuma informação sobre os mesmos. O objetivo é dividir os dados (vetores d-dimensionais) em grupos, de modo a que elementos do mesmo grupo sejam mais semelhantes entre si do que com membros de outros grupos. \
\
Desta forma pode-se categorizar os dados e descobrir as “classes” que neles estão subjacentes, obtendo assim uma representação de alto nível dos dados - encontrando-se estes agrupados em nuvens multi-dimensionais de pontos.

<a id='section7-3-2'></a>
### ◽ 3.2 Procura do Número de *Clusters* Ideal
Para estes testes, usar-se-á apenas uma quantidade reduzida de documentos - a quantidade de clusters seria, de outra forma, demasiado elevada.

Cria-se uma função que guarde o SSE - Sum Squared Error. A qualidade dos *clusters* criados é avaliada pela computação deste valor quando so centróides convergem ou quando igualam os resultados da iteração anterior. É definido como a soma dos quadrados das distâncias Euclideanas de cada ponto ao seu centro.

Um método popular de escolha de melhor número de *Clusters* é o chamado "método do cotovelo" - o método consiste em traçar o SSE em função do número de clusters e escolher "o cotovelo" da curva como o número de clusters a utilizar. No entanto, este método não é infalível - muitas vezes, com elevadas quantidades de dados desorganizados pode dificultar a convergência, impossibilitando a escolha de um cotovelo exato. Tal como se verifica no gráfico que segue:

```python
plt.plot(np.arange(20, 501, 20), pickle.load(open('resultados_clustering.p', 'rb'))['sse'], marker='o')
```
